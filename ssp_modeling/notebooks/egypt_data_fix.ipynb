## Fix "ww_domestic_rural" fracs in WALI

```python
ww_domestic_rural_cols_egypt = [col for col in egypt_df.columns if "ww_domestic_rural" in col and "frac_wali" in col]
ww_domestic_rural_cols_libya = [col for col in libya_df.columns if "ww_domestic_rural" in col and "frac_wali" in col]
ww_domestic_rural_cols_egypt.sort()
ww_domestic_rural_cols_libya.sort()

print(ww_domestic_rural_cols_egypt==ww_domestic_rural_cols_libya)
print(len(ww_domestic_rural_cols_egypt))
print(len(ww_domestic_rural_cols_libya))
```

```python
egypt_df[ww_domestic_rural_cols_egypt].sum(axis=1)
```

```python
libya_df[ww_domestic_rural_cols_libya].sum(axis=1)
```

```python
# update the egypt df with the libya values
egypt_df[ww_domestic_rural_cols_egypt] = libya_df[ww_domestic_rural_cols_libya].copy()
egypt_df[ww_domestic_rural_cols_egypt].sum(axis=1)
```
